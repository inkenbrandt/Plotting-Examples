# Matplotlib Tutorial for Beginners

This notebook covers the basics of creating plots with Matplotlib. We'll learn how to create simple plots with just a few commands and explore different types of visualizations.

## 1. Initialize
First, we need to import the required libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Create a Simple Plot
Let's create a basic sine wave plot:

```python
# Prepare the data
X = np.linspace(0, 10*np.pi, 1000)
Y = np.sin(X)

# Create the plot
fig, ax = plt.subplots()
ax.plot(X, Y)
plt.show()
```

## 3. Different Types of Plots
Matplotlib offers several kinds of plots. Let's explore some common ones:

```python
# Create a figure with 3x3 subplots
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

# 1. Scatter Plot
X = np.random.uniform(0, 1, 100)
Y = np.random.uniform(0, 1, 100)
axes[0, 0].scatter(X, Y)
axes[0, 0].set_title('Scatter Plot')

# 2. Bar Plot
X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
axes[0, 1].bar(X, Y)
axes[0, 1].set_title('Bar Plot')

# 3. Image Plot
Z = np.random.uniform(0, 1, (8, 8))
axes[0, 2].imshow(Z)
axes[0, 2].set_title('Image Plot')

# 4. Contour Plot
axes[1, 0].contourf(Z)
axes[1, 0].set_title('Contour Plot')

# 5. Pie Chart
Z = np.random.uniform(0, 1, 4)
axes[1, 1].pie(Z)
axes[1, 1].set_title('Pie Chart')

# 6. Histogram
Z = np.random.normal(0, 1, 100)
axes[1, 2].hist(Z)
axes[1, 2].set_title('Histogram')

# 7. Error Bar Plot
X = np.arange(5)
Y = np.random.uniform(0, 1, 5)
axes[2, 0].errorbar(X, Y, Y/4)
axes[2, 0].set_title('Error Bar Plot')

# 8. Box Plot
Z = np.random.normal(0, 1, (100, 3))
axes[2, 1].boxplot(Z)
axes[2, 1].set_title('Box Plot')

# 9. Line Plot with Styling
X = np.linspace(0, 10, 100)
Y = np.sin(X)
axes[2, 2].plot(X, Y, color='black', linestyle='--', linewidth=2, marker='o', markersize=4)
axes[2, 2].set_title('Styled Line Plot')

plt.tight_layout()
plt.show()
```

## 4. Customizing Plots
Let's explore different ways to customize our plots:

```python
# Create sample data
X = np.linspace(0, 10, 100)
Y1 = np.sin(X)
Y2 = np.cos(X)

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot with various customizations
ax.plot(X, Y1, label='Sine', color='blue', linestyle='-', linewidth=2)
ax.plot(X, Y2, label='Cosine', color='red', linestyle='--', linewidth=2)

# Add labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
ax.set_title('Sine and Cosine Waves')

# Add grid
ax.grid(True, linestyle=':')

# Add legend
ax.legend()

plt.show()
```

## 5. Saving Figures
You can save your figures in various formats:

```python
# Save as PNG with high DPI (dots per inch)
fig.savefig('my-first-figure.png', dpi=300)

# Save as PDF (vector format)
fig.savefig('my-first-figure.pdf')
```
