# Matplotlib Intermediate Guide

This notebook demonstrates intermediate-level Matplotlib concepts and techniques. We'll explore various visualization features and best practices for creating publication-ready figures.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator as ML
from matplotlib.ticker import ScalarFormatter as SF

# Enable inline plotting
%matplotlib inline
# Set figure style
#plt.style.use('seaborn-v0_8-whitegrid')
```

## 1. Advanced Tick Formatting

Learn how to customize tick locations, labels, and rotations for better readability.

```python
fig, ax = plt.subplots(figsize=(10, 6))
x = np.linspace(0, 4, 100)
y = np.sin(2 * np.pi * x)
ax.plot(x, y)

# Set minor ticks and format them
ax.xaxis.set_minor_locator(ML(0.2))
ax.xaxis.set_minor_formatter(SF())
ax.tick_params(axis='x', which='minor', rotation=90)

plt.title("Advanced Tick Formatting")
plt.show()
```

## 2. Lines and Markers

Explore various line styles and marker options for better data visualization.

```python
fig, ax = plt.subplots(figsize=(10, 6))
X = np.linspace(0.1, 10*np.pi, 1000)
Y = np.sin(X)

# Plot with custom markers and lines
ax.plot(X, Y, "C7o:", markevery=50, markeredgecolor="1.0", 
        label="Sine with markers")

plt.title("Custom Line and Marker Styles")
plt.legend()
plt.show()
```

## 3. Logarithmic Scales

Demonstrate how to work with different axis scales, particularly logarithmic scaling.

```python
fig, ax = plt.subplots(figsize=(10, 6))
X = np.logspace(-1, 2, 100)
Y = np.sin(X)

ax.set_xscale("log")
ax.plot(X, Y, "C7o-", markevery=10, markeredgecolor="1.0")

plt.title("Logarithmic Scale Example")
plt.show()
```

## 4. Text and Annotations

Learn how to add text, mathematical expressions, and annotations with arrows to your plots.

```python
fig, ax = plt.subplots(figsize=(10, 6))
X = np.linspace(0, 2*np.pi, 100)
Y = np.sin(X)

ax.plot(X, Y)
# Add period marking
ax.fill_betweenx((-1, 1), [0], [2*np.pi], alpha=0.1)
ax.text(np.pi, -1.2, r"Period $\Phi$", ha='center')

# Add annotation with arrow
ax.annotate("Peak", 
            xy=(np.pi/2, 1), 
            xytext=(np.pi/2, 0.5),
            ha="center",
            arrowprops={"arrowstyle": "->", "color": "C1"})

plt.title("Text and Annotations Example")
plt.show()
```

## 5. Multiple Plots with Custom Legend

Show how to combine multiple plots and customize legend placement.

```python
fig, ax = plt.subplots(figsize=(10, 6))
X = np.linspace(0, 4*np.pi, 100)

ax.plot(X, np.sin(X), "C0", label="Sine")
ax.plot(X, np.cos(X), "C1", label="Cosine")

# Custom legend placement
ax.legend(bbox_to_anchor=(0, 1, 1, .1), 
         ncol=2,
         mode="expand", 
         loc="lower left")

plt.title("Multiple Functions with Custom Legend")
plt.show()
```

## 6. Publication-Ready Figures

Learn how to create publication-quality figures with appropriate size and resolution.

```python
# Size for two-column A4 paper (3.75x3.75 inches)
fig = plt.figure(figsize=(3.75, 3.75))
ax = fig.add_subplot(111)

X = np.linspace(0, 5, 100)
ax.plot(X, np.sin(X))

plt.title("Publication-Ready Figure Size")
# Save with high DPI for print
plt.savefig("publication_figure.pdf", dpi=600)
plt.show()
```
