## 3. Figures and Axes

In Matplotlib, **Figures** are the windows or pages in which you place your plots, and **Axes** are the areas where the actual plots appear.

```python
# 3. FIGURES AND AXES

# Create a figure with a single Axes
fig, ax = plt.subplots()

# We can plot something trivial, say a simple list
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], marker='o')

# Set title and labels
ax.set_title("Basic Plot on Single Axes")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

plt.show()
```

By default, `plt.subplots()` creates one figure containing one Axes object (`ax`). You can also create multiple Axes in a grid.

## 4. Subplots

### 4.1 Multiple Subplots in One Figure

```python
# 4.1 MULTIPLE SUBPLOTS IN ONE FIGURE

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# axes is a 2D array of Axes objects
# Let's plot something on each of them

x = np.linspace(0, 2*np.pi, 100)

axes[0, 0].plot(x, np.sin(x), 'r-')
axes[0, 0].set_title('sin(x)')

axes[0, 1].plot(x, np.cos(x), 'g--')
axes[0, 1].set_title('cos(x)')

axes[1, 0].plot(x, np.sin(2*x), 'b-.')
axes[1, 0].set_title('sin(2x)')

axes[1, 1].plot(x, np.cos(2*x), 'm:')
axes[1, 1].set_title('cos(2x)')

# Adjust layout
plt.tight_layout()
plt.show()
```

Notice how we use an array of Axes objects to manage multiple subplots. Each subplot can be configured independently.

## 5. Basic Plot Types

### 5.1 Line Plot

```python
# 5.1 LINE PLOT

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()

# Create a line plot
ax.plot(x, y, label='sin(x)')

ax.set_title("Line Plot")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.legend()

plt.show()
```

### 5.2 Scatter Plot

```python
# 5.2 SCATTER PLOT

# Generate some random data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

fig, ax = plt.subplots()

# Create a scatter plot
scatter = ax.scatter(x, y, c=np.random.rand(100), cmap='viridis', alpha=0.7)

ax.set_title("Scatter Plot")
ax.set_xlabel("Random X")
ax.set_ylabel("Random Y")

plt.colorbar(scatter, ax=ax, label='Color scale')
plt.show()
```

### 5.3 Bar Chart

```python
# 5.3 BAR CHART

labels = ['A', 'B', 'C', 'D', 'E']
values = [10, 24, 36, 18, 12]

fig, ax = plt.subplots()

ax.bar(labels, values, color='skyblue')
ax.set_title("Bar Chart")
ax.set_ylabel("Value")

plt.show()
```

### 5.4 Pie Chart

```python
# 5.4 PIE CHART

sizes = [40, 35, 15, 10]
labels = ['Cats', 'Dogs', 'Rabbits', 'Other']

fig, ax = plt.subplots()

ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
ax.set_title("Pet Ownership")

plt.show()
```

### 5.5 Histogram

```python
# 5.5 HISTOGRAM

data = np.random.randn(1000)

fig, ax = plt.subplots()

# Plot a histogram of normally distributed data
ax.hist(data, bins=30, color='green', alpha=0.7, edgecolor='black')
ax.set_title("Histogram of Random Data")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

plt.show()
```

## 6. Legends, Labels, and Annotations

### 6.1 Adding Multiple Lines and a Legend

```python
# 6.1 MULTIPLE LINES AND LEGEND

x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()

ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cos(x)')

ax.set_title("Sine vs Cosine")
ax.set_xlabel("x")
ax.set_ylabel("Function value")
ax.legend(loc='upper right')

plt.show()
```

### 6.2 Text and Annotations

```python
# 6.2 TEXT AND ANNOTATIONS

fig, ax = plt.subplots()
x = np.linspace(0, 5, 100)
y = x**2

ax.plot(x, y, label='y = x^2')

# Add a text label at a specific position
ax.text(2, 10, "Quadratic Curve", fontsize=12, color='red')

# Annotate a point
ax.annotate('Max here', xy=(5, 25), xytext=(3, 30),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_title("Text and Annotations")
ax.legend()

plt.show()
```

## 7. Controlling Figure Layout

### 7.1 `plt.tight_layout()`

Using `plt.tight_layout()` automatically adjusts subplot parameters to give specified padding. We already used it once above. Lets see another example.

```python
# 7.1 TIGHT LAYOUT EXAMPLE

fig, axes = plt.subplots(2, 2, figsize=(8, 6))

for i, ax in enumerate(axes.flat, start=1):
    ax.plot(np.random.rand(10), np.random.rand(10), marker='o', linestyle='--')
    ax.set_title(f"Plot {i}")

# If we don't use tight_layout, titles may overlap
plt.tight_layout()
plt.show()
```

### 7.2 Adjusting Spacing Manually

```python
# 7.2 MANUAL LAYOUT ADJUSTMENT

fig, axes = plt.subplots(2, 2, figsize=(8, 6))
for i, ax in enumerate(axes.flat, start=1):
    ax.plot(np.random.rand(10), np.random.rand(10), marker='s')
    ax.set_title(f"Plot {i}")

# Adjust spacing manually
plt.subplots_adjust(wspace=0.3, hspace=0.4)
plt.show()
```

## 8. Colors, Markers, and Lines

Matplotlib offers a variety of ways to style your plots. You can customize **colors**, **markers**, and **line styles**.

```python
# 8. COLORS, MARKERS, AND LINES

x = np.linspace(0, 10, 10)

fig, ax = plt.subplots()

# Different line styles
ax.plot(x, x, color='red', linestyle='-', marker='o', label='Red Solid')
ax.plot(x, x+5, color='blue', linestyle='--', marker='^', label='Blue Dashed')
ax.plot(x, x+10, color='green', linestyle=':', marker='s', label='Green Dotted')

ax.set_title("Colors, Markers, and Lines")
ax.legend()

plt.show()

# A few helpful references:
# - Colors: 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w', or any HTML hex color code.
# - Markers: '.', 'o', '^', 's', '*', etc.
# - Line styles: '-', '--', ':', '-.'
```

## 9. Saving Figures

You can save figures by calling `plt.savefig("filename.ext")`.

```python
# 9. SAVING FIGURES

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, label='sin(x)')
ax.set_title("Saving Figures Example")
ax.legend()

# Save to a PNG file
plt.savefig("example_sin_wave.png", dpi=150, bbox_inches='tight')

plt.show()
```

This will create a file named `example_sin_wave.png` in your working directory.

## 10. Displaying Images

Matplotlib can also display images using `imshow`.

```python
# 10. DISPLAYING IMAGES

# Create a 10x10 random 'image'
image_data = np.random.rand(10, 10)

fig, ax = plt.subplots()
cax = ax.imshow(image_data, cmap='viridis')

ax.set_title("Random Image")
fig.colorbar(cax)

plt.show()
```

## 11. 3D Plots

Matplotlibs `mpl_toolkits.mplot3d` toolkit lets you create 3D plots.

```python
# 11. 3D PLOTS

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(projection='3d')  # or: ax = fig.add_subplot(111, projection='3d')

# Create data
theta = np.linspace(-4*np.pi, 4*np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='3D Spiral')
ax.set_title("3D Plot Example")
ax.legend()

plt.show()
```
