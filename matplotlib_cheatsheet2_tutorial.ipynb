# Matplotlib Examples Tutorial
This notebook demonstrates many of the core functionalities of [**Matplotlib**](https://matplotlib.org/), inspired by the [Matplotlib Cheat Sheet (Part 2)](https://matplotlib.org/cheatsheets/_images/cheatsheets-2.png). It covers creating plots, customizing them, and using various plot types.

> **Instructions**: Run each cell (Shift + Enter) to see the output and follow along with the comments.

```python
# We start by importing the necessary libraries.
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# For reproducibility:
np.random.seed(42)

print("Numpy version:", np.__version__)
print("Matplotlib version:", matplotlib.__version__)
```

## 1. Basic Setup
### 1.1 Creating a Simple Figure and Axes
In Matplotlib, you can create a figure and axis (or multiple axes) in many ways. One common approach is `plt.subplots()`.

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [1, 4, 9])  # Simple line plot
ax.set_title("Basic Line Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.show()
```

## 2. Figure Management and Axes
### 2.1 Figure Size, DPI, and Layout
You can customize the size of the figure and its resolution (DPI) by passing parameters to `plt.subplots()` or `plt.figure()`.

```python
# Example with figsize and dpi
fig, ax = plt.subplots(figsize=(6, 4), dpi=120)
ax.plot([1, 2, 3], [2, 4, 3], label='Test line')
ax.set_title("Customized Figure Size and DPI")
ax.legend()
plt.show()
```

### 2.2 Multiple Axes (Subplots)
To create multiple plots in a single figure, you can use either the `subplot` function or `subplots(nrows, ncols)`.

```python
# 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# We'll fill these plots with random data for demonstration
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

axs[0,0].plot(x, y_sin, color='blue')
axs[0,0].set_title("Sine")

axs[0,1].plot(x, y_cos, color='red')
axs[0,1].set_title("Cosine")

axs[1,0].scatter(x, y_sin, color='green')
axs[1,0].set_title("Scatter Sine")

axs[1,1].scatter(x, y_cos, color='magenta')
axs[1,1].set_title("Scatter Cosine")

fig.tight_layout()
plt.show()
```

## 3. Different Plot Types
### 3.1 Line Plot
A classic line plot can be created using either `plt.plot()` or `ax.plot()`.

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label='Sine Wave', color='green', linewidth=2, linestyle='--')
ax.set_title("Line Plot Example")
ax.set_xlabel("X values")
ax.set_ylabel("Sine of X")
ax.legend()
plt.show()
```

### 3.2 Scatter Plot
Scatter plots are useful for showing relationships between two sets of data.

```python
fig, ax = plt.subplots()
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 200 * np.random.rand(50)

scatter = ax.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.6)
ax.set_title("Scatter Plot Example")
ax.set_xlabel("Random X")
ax.set_ylabel("Random Y")

# Add a colorbar to show the color scale
cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label("Color Scale")

plt.show()
```

### 3.3 Bar Plot
Bar plots are often used to display discrete data or categories.

```python
labels = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

fig, ax = plt.subplots()
ax.bar(labels, values, color='orange', edgecolor='black')
ax.set_title("Bar Plot Example")
ax.set_xlabel("Categories")
ax.set_ylabel("Values")
plt.show()
```

#### 3.3.1 Horizontal Bar Plot
A horizontal bar plot can be created with `barh()`.

```python
fig, ax = plt.subplots()
ax.barh(labels, values, color='skyblue', edgecolor='black')
ax.set_title("Horizontal Bar Plot Example")
ax.set_xlabel("Values")
ax.set_ylabel("Categories")
plt.show()
```

### 3.4 Histogram
Histograms are used to display distributions of numeric data.

```python
data = np.random.randn(1000)  # 1000 random points from a normal distribution

fig, ax = plt.subplots()
ax.hist(data, bins=30, color='purple', alpha=0.7)
ax.set_title("Histogram Example")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")
plt.show()
```

### 3.5 Pie Chart
Pie charts display the proportion of each category in a dataset.

```python
fig, ax = plt.subplots()
pie_values = [10, 20, 30, 40]
pie_labels = ['Apple', 'Banana', 'Cherry', 'Date']

ax.pie(pie_values, labels=pie_labels, autopct='%1.1f%%', startangle=140)
ax.set_title("Pie Chart Example")
plt.show()
```

### 3.6 Box Plot (Box and Whiskers)
Box plots (a.k.a. box-and-whisker plots) display the quartiles of the data.

```python
# Generate random data for demonstration
data_for_boxplot = np.random.rand(4, 50)  # 4 groups, 50 data points each

fig, ax = plt.subplots()
ax.boxplot(
    data_for_boxplot.T,
    patch_artist=True, 
    boxprops=dict(facecolor='lightblue', color='blue'),
    medianprops=dict(color='red')
)
ax.set_title("Box Plot Example")
ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(['Group1', 'Group2', 'Group3', 'Group4'])
plt.show()
```

### 3.7 Violin Plot
Violin plots are similar to box plots, but also show the probability density.

```python
fig, ax = plt.subplots()
ax.violinplot(data_for_boxplot, showmeans=False, showmedians=True)
ax.set_title("Violin Plot Example")
ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(['Group1', 'Group2', 'Group3', 'Group4'])
plt.show()
```

## 4. Styling and Customization
### 4.1 Changing Colors, Markers, and Line Styles
Matplotlib allows you to style your lines with markers, different colors, and different line styles.

```python
fig, ax = plt.subplots()
x = np.linspace(0, 5, 10)
ax.plot(x, x, color='blue', marker='o', linestyle='-', label='Line 1')
ax.plot(x, x**2, color='red', marker='s', linestyle='--', label='Line 2')
ax.plot(x, x**3, color='green', marker='^', linestyle='-.', label='Line 3')
ax.set_title("Line Styles and Markers")
ax.legend()
plt.show()
```

### 4.2 Adding Annotations
You can add annotations (arrows, text) to highlight specific points of interest.

```python
fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
ax.plot(x, y, color='darkcyan')

# Annotate the peak
peak_x = np.pi/2
peak_y = 1
ax.annotate(
    'Maximum',
    xy=(peak_x, peak_y),
    xytext=(peak_x + 0.5, peak_y + 0.3),
    arrowprops=dict(facecolor='black', shrink=0.05),
    fontsize=12
)
ax.set_title("Annotation Example")
plt.show()
```

### 4.3 Setting Ticks and Tick Labels
You can set custom ticks and labels on the axes:

```python
fig, ax = plt.subplots()
x = np.arange(0, 10, 1)
y = x**2
ax.plot(x, y)

# Set x ticks at every integer
ax.set_xticks(x)
# Set custom labels for demonstration
ax.set_xticklabels(['Zero','One','Two','Three','Four','Five','Six','Seven','Eight','Nine'])

ax.set_title("Custom Ticks and Labels")
plt.show()
```

### 4.4 Log Scales
For data spanning several orders of magnitude, a logarithmic scale can be used.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))

x = np.linspace(0.1, 100, 100)
y = x**2

# Linear scale
ax1.plot(x, y)
ax1.set_title("Linear Scale")

# Log scale
ax2.plot(x, y)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_title("Log Scale")

plt.tight_layout()
plt.show()
```

## 5. Subplot Adjustments and Layouts
### 5.1 `tight_layout()`
`tight_layout()` automatically adjusts subplot parameters so that subplots fit nicely into the figure area.

```python
fig, axs = plt.subplots(2, 2)

for i, ax in enumerate(axs.flat):
    ax.plot(np.random.rand(10), np.random.rand(10), marker='o')
    ax.set_title(f"Plot {i+1}")

plt.tight_layout()
plt.show()
```

### 5.2 `GridSpec` for Advanced Layout
If you want more control over subplot sizes and arrangement, `GridSpec` can be used.

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(6, 6))
gs = gridspec.GridSpec(3, 3, figure=fig)

# Large top-left plot spanning rows 0,1 and columns 0,1
ax1 = fig.add_subplot(gs[0:2, 0:2])
ax1.plot(np.random.rand(50), np.random.rand(50), 'o')
ax1.set_title("Large Plot")

# Smaller plots
ax2 = fig.add_subplot(gs[0, 2])
ax3 = fig.add_subplot(gs[1, 2])
ax4 = fig.add_subplot(gs[2, 0:2])
ax5 = fig.add_subplot(gs[2, 2])

ax2.plot(np.random.randn(50).cumsum())
ax2.set_title("Small 1")

ax3.plot(np.random.randn(50).cumsum())
ax3.set_title("Small 2")

ax4.plot(np.random.randn(50).cumsum())
ax4.set_title("Wide Plot")

ax5.plot(np.random.randn(50).cumsum())
ax5.set_title("Small 3")

plt.tight_layout()
plt.show()
```

## 6. Text and Label Customization
### 6.1 Title, Axis Labels, and Legend

```python
fig, ax = plt.subplots()
x = np.linspace(0, 4*np.pi, 200)
y = np.sin(x)

ax.plot(x, y, label='Sine wave')
ax.set_xlabel("X Values")
ax.set_ylabel("sin(X)")
ax.set_title("Title Example with Legend")
ax.legend(loc='upper right')  # Move legend to upper right corner
plt.show()
```

### 6.2 Font Size and Style

```python
fig, ax = plt.subplots()
ax.plot(x, y, label='Sine wave')

# Customize fonts
ax.set_xlabel("X Values", fontsize=14, fontweight='bold')
ax.set_ylabel("sin(X)", fontsize=14, fontstyle='italic')
ax.set_title("Font Style Example", fontsize=16, fontweight='bold')
ax.legend(loc='upper right', fontsize=12)
plt.show()
```

You can save figures in multiple formats (PNG, PDF, SVG, etc.) using `plt.savefig()` or `fig.savefig()`.

## 8. Putting It All Together
Below is a more advanced, combined example that shows multiple plot types, styling, and saving:

```python
import matplotlib.gridspec as gridspec  # Re-import in case this cell runs standalone

fig = plt.figure(figsize=(8,6))
gs = gridspec.GridSpec(2,2, figure=fig)

x = np.linspace(0, 10, 100)

# 1. Line plot
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(x, np.sin(x), label='sin(x)', color='navy')
ax1.set_title("Line Plot")
ax1.legend()

# 2. Scatter plot
ax2 = fig.add_subplot(gs[0,1])
ax2.scatter(x, np.cos(x), color='crimson')
ax2.set_title("Scatter Plot")

# 3. Histogram
ax3 = fig.add_subplot(gs[1,0])
random_data = np.random.randn(1000)
ax3.hist(random_data, bins=20, color='darkgreen', alpha=0.7)
ax3.set_title("Histogram")

# 4. Pie chart
ax4 = fig.add_subplot(gs[1,1])
pie_vals = [15, 30, 45, 10]
pie_labels = ['A', 'B', 'C', 'D']
ax4.pie(pie_vals, labels=pie_labels, autopct='%.0f%%', startangle=140)
ax4.set_title("Pie Chart")

plt.tight_layout()

# Example of saving:
# fig.savefig("combined_example.png", dpi=300)
plt.show()
```
